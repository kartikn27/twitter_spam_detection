```python
def get_tweet(id):
    try:
        tweet = twitter.show_status(id = id)
        
        if tweet:
            data = {}
            data["created_at"] = tweet["created_at"]
            data["id"] = tweet["id"]
            data["text"] = tweet["text"]
            data["user_id"] = tweet["user"]["id"]
            data["user_name"] = tweet["user"]["name"]
            data["user_location"] = tweet["user"]["location"]
            data["user_description"] = tweet["user"]["description"]
            data["hashtags"] = tweet["entities"]["hashtags"]
            data["symbols"] = tweet["entities"]["symbols"]
            data["urls"] = tweet['entities']["urls"]
            data["followers_count"] = tweet["user"]["followers_count"]
            data["friends_count"] = tweet["user"]["friends_count"]
            data["verified"] = tweet["user"]["verified"]                
            
#             return(tweet['text'])
        data_arr.append(data)
        
    except TwythonError as e:
        if e.error_code == 403:
            pass
```

```python
t_id['tweet'] = t_id["id"].apply(get_tweet)
```
